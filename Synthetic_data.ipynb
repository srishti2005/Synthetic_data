```python
from random import randint
def imput_data(x):
  #pandas dataframe
  data = pd.DataFrame()
  for i in range(0,x):
    data.loc[i,'id'] = randint(1,100)
    data.loc[i,'name'] = fake.name()
    data.loc[i,'address'] = fake.address()
    data.loc[i,'latitude'] = fake.latitude()
    data.loc[i,'longitude'] = fake.longitude()
  return data
imput_data(10)
```

```python
# A list of r features
features =[
    "id",
    "gender",
    "subscriber",
    "name",
    "rating"
]

#Creating a DF for these features
df = pd.DataFrame(columns=features)
df
```

RSeeding restricts the generation of random entities

```python
import numpy as np
Faker.seed(42)
def generate_persons(num =1):
  fake = Faker()
  rows = [{
      'person_id': fake.uuid4(),
      'first_name': fake.first_name(),
      'last_name': fake.last_name(),
      'address': fake.address(),
      'ssn': fake.ssn(),
      'dob': fake.date_of_birth(minimum_age=18,maximum_age=75)

  }
  for x in range(num)]
  return pd.DataFrame(rows)
persons=generate_persons(1000)
print(persons)
```

```python
num_users = 100
df['id']=[uuid.uuid4().node for i in range(num_users)]
df
```

```python
print(df['id'].nunique() == num_users)
```

```python
genders = ["male","female", "NA"]
df['gender'] = random.choices(genders,weights=(45,45,1), k=num_users)
df
```

Gives the counts of the number of persons in each gender

```python
df.gender.value_counts()
```

```python
#choices
choice = [True,False]
df['subscribers']= random.choices(
    choice,
    k=num_users
)
df
```

```python
#Initialising Faker
faker = Faker(locale="FR_FR")
def name_gen(gender):
  if gender == "male":
    return faker.name_male()
  elif gender == "female":
    return faker.name_female()
  return faker.name

df['name'] = [name_gen(i) for i in df['gender']]
df

```

```python
#rating
rating = [1,2,3,4,5]
df['rating']= random.choices(
    rating,
    k=num_users
)
df
```

convert dataframe to dataset
